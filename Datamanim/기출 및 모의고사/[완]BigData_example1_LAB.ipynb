#### 2-2. 하우징 데이터
- housing.csv 파일을 사용한다
- 주어진 데이터의 첫번째 행 부터 순서대로 80% 까지의 데이터를 추출 후 'total_bedrooms' 변수의 결측값(NA)을 'total_bedrooms' 변수의 중앙값으로 대체하고,
- 대체 전의 'total_bedrooms' 변수 표준편차값과 대체 후의 'total_bedrooms' 변수 표준편차 값을 산출하여
- 대체 전 표준편차 값 - 대체 후 표준편차 값을 구해 출력한다.

```python
origin = pd.read_csv('./bigdata/housing.csv')
before = origin[:int(len(origin)*0.8)]
before_std = before['total_bedrooms'].std()

after = before.copy()

after['total_bedrooms'] = after['total_bedrooms'].fillna(after['total_bedrooms'].median())
after_std = after['total_bedrooms'].std()

before_std - after_std
```

#### 2-3. 하우징 데이터 
- 'https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/housing.csv' 파일을 사용한다
- 하우징 데이터에서 'latitude' 컬럼의 이상치를 찾아 이상치들의 합을 산출하시오.
- 이상치 기준 : 평균 - (표준편차 * 1.5), 평균 + (표준편차 * 1.5) 

```python
df = pd.read_csv('./bigdata/housing.csv')
t = df['latitude'].std()*1.5
mean = df['latitude'].mean()

top = mean + t
bot = mean - t

a = df[df['latitude'] >= top].latitude.sum()
b = df[df['latitude'] <= bot].latitude.sum()
a+b
```

```python
# [0] 사용 라이브러리 import
import pandas as pd
import numpy as np

# 데이타가 많은경우 모두 출력 안되고 ... 으로 생략해서 출력됨.
pd.set_option('display.max_rows',500)    #출력할 max row를 지정
pd.set_option('display.max_columns',20)  #출력할 max columns를 지정
#출력 format 지정 - 소수점아래 4자리까지
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
# [0] x_train, x_test 데이터로 생성함
# 이 부분은 시험에서는 없습니다. 그냥 데이터를 3개로 나누기 위해 사용한 것입니다.
from sklearn.model_selection import train_test_split

dftot = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/1st_Train.csv')
x_train, x_test = train_test_split(dftot, test_size=0.4, stratify=dftot['Reached.on.Time_Y.N'], random_state=0)
y_train = x_train[['Reached.on.Time_Y.N']]
X_train = x_train.drop(columns='Reached.on.Time_Y.N')
y_test = x_test[['Reached.on.Time_Y.N']]   # 이것은 시험에서는 제공되지 않음
X_test = x_test.drop(columns='Reached.on.Time_Y.N')

X_train.to_csv('x_train.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
X_test.to_csv('x_test.csv', index=False)
y_test.to_csv('y_test.csv', index=False) # 이것은 시험에서는 제공되지 않음
```

```python
X_train = X_train.drop(columns = 'ID')
X_test_ID = X_test.pop('ID')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X_train['Warehouse_block'] = le.fit_transform(X_train['Warehouse_block'])
X_train['Mode_of_Shipment'] = le.fit_transform(X_train['Mode_of_Shipment'])
X_train['Product_importance'] = le.fit_transform(X_train['Product_importance'])
X_train['Gender'] = le.fit_transform(X_train['Gender'])

X_test['Warehouse_block'] = le.fit_transform(X_test['Warehouse_block'])
X_test['Mode_of_Shipment'] = le.fit_transform(X_test['Mode_of_Shipment'])
X_test['Product_importance'] = le.fit_transform(X_test['Product_importance'])
X_test['Gender'] = le.fit_transform(X_test['Gender'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

ls = ['Cost_of_the_Product', 'Weight_in_gms', 'Discount_offered']

X_train[ls] = scaler.fit_transform(X_train[ls])
X_test[ls] = scaler.fit_transform(X_test[ls])
```

```python
display(X_train)
display(y_train)
display(X_test)
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X_train, y_train, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

modelRF = RandomForestClassifier(random_state = 1234)
modelRF.fit(X_tr, y_tr)
modelRF.score(X_ts, y_ts)
```

```python
tmp = modelRF.predict(X_ts)
y_pred = modelRF.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
```

```python
print(roc_auc_score(tmp, y_ts))
print(f1_score(tmp, y_ts))
```

```python
print(roc_auc_score(y_pred, y_test))
print(f1_score(y_pred, y_test))
```
