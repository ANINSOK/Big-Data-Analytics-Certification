## 데이터 불러오기

```python
import pandas as pd
#데이터 로드
X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/y_train.csv")
X_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/x_test.csv")
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/smoke/y_test.csv')


display(X_train.head())
display(X_test.head())
display(y_train.head())
display(y_test.head())
```

### isna

### fillna

```python
X_train = X_train.drop(columns = 'ID')
y_train = y_train.drop(columns = 'ID')
X_test_ID = X_test.pop('ID')
```

### 표준화

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
X_train[['수축기혈압', '트리글리세라이드', '(혈청지오티)ALT']] = Scaler.fit_transform(X_train[['수축기혈압', '트리글리세라이드', '(혈청지오티)ALT']])
X_test[['수축기혈압', '트리글리세라이드', '(혈청지오티)ALT']] = Scaler.fit_transform(X_test[['수축기혈압', '트리글리세라이드', '(혈청지오티)ALT']])
```

### 원핫인코딩

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

### 검증용 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X_train, y_train, test_size=0.3)
```

## 모델링

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(random_state=1234)
modelRF.fit(X_tr, y_tr)
modelRF.score(X_ts, y_ts)
```

```python
y_pred = modelRF.predict(X_test)
```

```python
a = pd.DataFrame({'ID':X_test_ID, '흡연상태':y_pred})
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(a['흡연상태'], y_test['흡연상태'])
```
