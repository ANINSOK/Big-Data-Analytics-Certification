## 데이터 불러오기

```python
import pandas as pd
#데이터 로드
X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/y_train.csv")
X_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_test.csv")
y_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/y_test.csv")


display(X_train.head())
display(y_train.head())
display(X_test.head())
```

```python
print(X_train.shape)
print('-'*80)
print(X_test.shape)
print('-'*80)
print(y_train.shape)
print('-'*80)

print(X_train.info())
print('-'*80)
print(X_test.info())
print('-'*80)
print(y_train.info())
print('-'*80)

print(X_train.describe())
print('-'*80)
print(X_test.describe())
print('-'*80)
print(y_train.describe())
print('-'*80)
```

```python
print(X_train.isna().sum().sum())
print('-'*80)
print(X_test.isna().sum().sum())
print('-'*80)
print(y_train.isna().sum().sum())
print('-'*80)
```

```python
X_train = X_train.drop(columns = 'ID')
y_train = y_train.drop(columns = 'ID')
X_test_ID = X_test.pop('ID')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X_train['model'] = le.fit_transform(X_train['model'])
X_train['fuelType'] = le.fit_transform(X_train['fuelType'])
X_train['transmission'] = le.fit_transform(X_train['transmission'])

X_test['model'] = le.fit_transform(X_test['model'])
X_test['fuelType'] = le.fit_transform(X_test['fuelType'])
X_test['transmission'] = le.fit_transform(X_test['transmission'])
```

```python
display(X_train.head())
display(y_train.head())
display(X_test.head())

print(X_train.describe())
print('-'*80)
print(X_test.describe())
print('-'*80)
print(y_train.describe())
print('-'*80)
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X_train, y_train, test_size=0.3)

from sklearn.ensemble import RandomForestRegressor
modelRF = RandomForestRegressor()
modelRF.fit(X_tr, y_tr)
modelRF.score(X_ts, y_ts)
```

```python
result = modelRF.predict(X_ts)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
print(mean_absolute_error(y_ts, result))
MES = mean_squared_error(y_ts, result)
print(MES)
print(np.sqrt(MES))
print(r2_score(y_ts, result))
```

```python
y_pred = modelRF.predict(X_test)
```

```python
print(mean_absolute_error(y_pred, y_test['price']))
MES = mean_squared_error(y_pred, y_test['price'])
print(MES)
print(np.sqrt(MES))
print(r2_score(y_pred, y_test['price']))
```

```python
pd.DataFrame({'ID':X_test_ID, 'price':y_pred}).to_csv('회귀 7. 현대 차량 가격 분류문제 데이터.csv', index=False)
```
