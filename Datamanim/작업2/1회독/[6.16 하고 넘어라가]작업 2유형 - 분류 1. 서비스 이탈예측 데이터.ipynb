## 데이터 불러오기

```python
import pandas as pd
#데이터 로드
X_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
X_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")



display(X_train.head())
display(X_test.head())
display(y_train.head())
```

## 1. 데이터셋 확인, EDA

### shape

```python
print(X_train.shape)
print("-"*80)
print(y_train.shape)
print("-"*80)
print(X_test.shape)
print("-"*80)
```

### info

```python
print(X_train.info())
print("-"*80)
print(y_train.info())
print("-"*80)
print(X_test.info())
print("-"*80)
```

### isna

```python
print(X_train.isna().sum())
print("-"*80)
print(y_train.isna().sum())
print("-"*80)
print(X_test.isna().sum())
print("-"*80)
```

### fillna

```python
X_train = X_train.drop(columns=['Surname','CustomerId'])
X_test = X_test.drop(columns=['Surname','CustomerId'])
y_train = y_train.drop(columns=['CustomerId'])
```

## 2. Feature Engineering

### 표준화

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()

X_train[['CreditScore','Age', 'Balance','EstimatedSalary']] = Scaler.fit_transform(X_train[['CreditScore','Age', 'Balance','EstimatedSalary']])
X_test[['CreditScore','Age', 'Balance','EstimatedSalary']] = Scaler.fit_transform(X_test[['CreditScore','Age', 'Balance','EstimatedSalary']])
```

### 원핫인코딩

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
display(X_train.head())
display(X_test.head())
display(y_train.head())
```

### 검증용 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X_train, y_train, test_size=0.3)
```

## 모델링

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier()
modelRF.fit(X_tr, y_tr)
modelRF.score(X_ts, y_ts)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = modelRF.predict(X_ts)
y_pred
```

```python
from sklearn.metrics import roc_auc_score
R_A_score=roc_auc_score(y_ts, y_pred)
R_A_score
```
