## 데이터 불러오기

```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

### 1. 각 데이터의 ‘ct’컬럼을 시간으로 인식할수 있게 datatype을 변경하고 video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라

```python
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])
video['videoname'].value_counts()
```

### 2. 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라

```python
video.sort_values('ct', ascending = False).drop_duplicates('videoname', keep = 'first')[['videoname', 'viewcnt' ,'ct']].sort_values('videoname').reset_index(drop=True)
```

### 3. Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라

```python
channel[channel['ct'] > pd.to_datetime('2021-10-03')].sort_values('ct').drop_duplicates('channelname', keep='first')[['channelname', 'subcnt']]
```

### 4. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
start = channel[channel['ct'] >= pd.to_datetime('2021-10-03 03:00:00')]
end = channel[channel['ct'] <= pd.to_datetime('2021-11-01 15:00:00')]
start = start.sort_values('ct').drop_duplicates('channelname', keep='first')
end = end.sort_values('ct', ascending = False).drop_duplicates('channelname', keep='first')
start = start.sort_values('channelname')[['channelname', 'subcnt']].reset_index(drop=True)
end = end.sort_values('channelname')[['channelname', 'subcnt']].reset_index(drop=True)
end['subcnt2'] = end['subcnt']
end = end.drop(columns = 'subcnt')

target = pd.merge(start, end)
target['diff'] = target['subcnt2'] - target['subcnt']
target[['channelname', 'diff']]
```

### 5. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다. 공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라

```python
# 알고리즘 처럼 풀어보고 싶은 문제임
ep1 = video[video['videopk'] == 'c5JQp6xafqc']
ep1 = ep1.sort_values('ct').reset_index(drop=True)
ep1['min'] = ep1['ct'].dt.minute
ep1.value_counts('min')
display(ep1[(ep1['min'] == 41) | (ep1['min'] == 13) | (ep1['min'] == 23)]) # 여기서 인덱스 번호 보고 일일이 리스트 만들었음 ㅎㅎ;;
ls = [720, 721, 722, 1624, 1625, 1626]
ep1[ep1.index.isin(ls)]
```

```python
# 이게 절대 맞는 풀이가 아니라는 생각이 듦
import datetime

ep_one = video.loc[video.videoname.str.contains('1')].sort_values('ct').reset_index(drop=True)

ep_one[
        (ep_one.ct.diff(1) >=datetime.timedelta(minutes=20)) | \
        (ep_one.ct.diff(1) <=datetime.timedelta(minutes=5))
      
      ]

answer = ep_one[ep_one.index.isin([720,721,722,723,1635,1636,1637])]
display(answer)
```

### 6. 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라

```python
target = video.sort_values('ct')
target['YMD'] =target['ct'].dt.date 
target.drop_duplicates('videopk', keep='first')[['videoname', 'YMD']]
```

### 7. “공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다. 공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
target = video.copy()
target['hour'] = target['ct'].dt.hour
target[target['hour'] == 21].sort_values('ct').drop_duplicates('videoname', keep='first')[['videoname','viewcnt','ct']].sort_values('viewcnt', ascending = False)
```

### 8. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

```python
target = video.copy()
target = target.sort_values('ct').drop_duplicates('videoname', keep='last')
target['ratio'] = target['dislikecnt']/target['likecnt']
target[['videoname', 'ratio']].sort_values('ratio')
```

### 9. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
target = video.copy()
target = target[(target['ct'] >= pd.to_datetime('2021-11-01 00:00:00')) & (target['ct'] <= pd.to_datetime('2021-11-01 15:00:00'))]
first = target.drop_duplicates('videoname', keep='first')[['videoname', 'viewcnt']]
last = target.drop_duplicates('videoname', keep='last')[['videoname', 'viewcnt']]
last.rename(columns = {'viewcnt':'viewcnt2'}, inplace = True)
ans = first.merge(last)
ans['diff'] = ans['viewcnt2']-ans['viewcnt']
ans[['videoname','diff']]
```

### 10. video 데이터 중에서 중복되는 데이터가 존재한다. 중복되는 각 데이터의 시간대와 videoname 을 구하여라

```python
video[video.duplicated()][['videoname', 'ct']]
```
